Predict Fear / Greed classification

```python
merged = pd.read_csv("ds_raina/csv_files/merged_trader.csv")

# Keep only rows with valid classification (Fear/Greed)
df = merged.dropna(subset=['classification'])
df = df[df['classification'].isin(['Fear', 'Greed'])]   # remove accidental other values

print("Rows after cleaning:", df.shape)
```

```python
features = ['execution_price','size_usd','fee','crossed']

X = df[features]
y = df['classification']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=200,max_depth=7,random_state=42)

model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print("Classification Accuracy:", accuracy)
```
